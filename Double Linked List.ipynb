## Double Linked List
### 1. Add element
    a. add element at beginning
    b. add element at end
    c. add element inbetween
    
### 2. Remove element
    a. remove element from beginning
    b. remove element from last
    c. remove element inbetween
### 3. Traversal

```python
class Node: 
    def __init__(self, data):
        self.data = data
        self.pref = None
        self.nref = None
        
class Double_Linked_List:
    def __init__(self):
        self.head = None
        
    # 3. Traversal   
    def traverse(self):
        if self.head is None:
            print('Linked List is Empty.!!')
        n = self.head
        while n is not None:
            print(f'{n.data}--->', end=" ")
            n = n.nref
            
    # 1.a. add element at beginning        
    def add_begin(self, data):
        if self.head is None:
            n = self.head
            new_node = Node(data)
            new_node.nref = n
            self.head = new_node
            return 
        n = self.head
        new_node = Node(data)
        new_node.nref = n
        n.pref = new_node
        self.head = new_node
    
    # 1.b. add element at end
    def add_end(self, data):
        new_node = Node(data)
        if self.head is None:
            print("Linked List is Empty.!!")
            return
        n = self.head
        while n.nref is not None:
            n = n.nref
        n.nref = new_node
        new_node.pref = n
        
    # 1.c. add element inbetween    
    def add_inbetween(self, data, x, y):
        new_node = Node(data)
        if self.head is None:
            print('Linked List is Empty.!!')
            return
        n = self.head
        while n.data !=x and n.nref.data !=y:
            n = n.nref
        new_node.pref = n
        new_node.nref = n.nref
        n.nref = new_node
        n.nref.pref = new_node
    
    # 2.a. remove element from beginning
    def delete_begin(self):
        if self.head is None:
            print('Linked List is Empty.!!')
            return
        n = self.head
        self.head = n.nref
     
    # 2.b. remove element from last
    def delete_end(self):
        if self.head is None:
            print('Linked List is Empty.!!')
            return
        if self.head.nref is None:
            self.head=None
            return
        n = self.head
        while n.nref.nref is not None:
            n = n.nref
        n.nref = n.nref.nref   
    
    # 3.c. remove element inbetween
    def delete_inbetween(self, x):
        if self.head is None:
            print('Linked List is Empty.!!')
            return 
        if self.head.data==x:
            n = self.head
            self.head = n.nref
            return
        n = self.head
        while n.nref.data != x:
            n = n.nref
        if n.nref.nref is not None:
            n.nref = n.nref.nref
            n.nref.pref = n
        else:
            n.nref = n.nref.nref

        
        
DLL = Double_Linked_List()
DLL.add_begin(50)
print(DLL.traverse())
DLL.add_end(352)
print(DLL.traverse())
DLL.add_begin(100)
print(DLL.traverse())
DLL.add_begin(5000)
print(DLL.traverse())
DLL.add_inbetween(415, 100, 50)
DLL.add_inbetween(1743, 50, 352)
print(DLL.traverse())
DLL.delete_begin()
print(DLL.traverse())
DLL.delete_end()
print(DLL.traverse())
DLL.delete_inbetween(50)
print(DLL.traverse())
```
